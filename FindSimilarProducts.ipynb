```python
product = r"C:\Users\mauro\Downloads\Migros_case\products\en"

import pandas as pd
import os
import json

def load_json_files(directory):
    products = []
    for filename in os.listdir(directory):
        if filename.endswith(".json"):
            with open(os.path.join(directory, filename), 'r') as f:
                try:
                    product_data = json.load(f)
                    products.append(product_data)
                except json.JSONDecodeError:
                    print(f"Couldn't parse {filename} as JSON.")
    return products
```

```python
import re

def remove_words_with_numbers(input_string):
    # Define a regular expression pattern to match words with numbers
    pattern = r'\b\w*\d\w*\b'
    
    # Use re.sub() to replace words with numbers with an empty string
    result_string = re.sub(pattern, '', input_string)
    
    return result_string

IDS = list(circ['Product_number'])
productls = [remove_words_with_numbers(str(sentence)) for sentence in list(circ['Product_description'])]
```

```python
#!pip install -U sentence-transformers

from sentence_transformers import SentenceTransformer, util
```

```python
sentences = ["I'm happy", "I'm full of happiness"]
# Multilingual sentence-transformer to generate embeddings
model = SentenceTransformer('sentence-transformers/distiluse-base-multilingual-cased-v1')

#Compute embedding for both lists
embedding_1= model.encode(sentences[0], convert_to_tensor=True)
embedding_2 = model.encode(sentences[1], convert_to_tensor=True)

util.pytorch_cos_sim(embedding_1, embedding_2)
## tensor([[0.6003]])
```

```python
sentences = productls
```

```python
#this is slow
embeddings = [model.encode(sentence) for sentence in sentences]
# PER MONO IMPORTANTE LEGGI STO COMMENTO, SE NON PRECOMPUTI GLI EMBEDDING E LI SALVI NON é FATTIILE
# then its fast: we need to store the embeddings in the database and on the fly only compute the cosine similarity
choosen = model.encode("torta", convert_to_tensor=True)
similarities = [util.pytorch_cos_sim(choosen, embedding) for embedding in embeddings[1:]]
# find max index in similarities list
idx = similarities.index(max(similarities))   
print(f"original was torta while most similar is {sentences[idx]}")
```

### FUNZIONA COSI COSA MA NON É TERRIBILE
Now we need to include the eco ratings, we only suggest if its more ecofriendly

```python
# Randomly pick a product
idx = 254
choosenID = IDS[idx]
# Compute embedding of chosen product
choosen = model.encode(sentences[idx], convert_to_tensor=True)
# compute similarities with other embeddings
embeddings = [arr for arr in embeddings if arr != choosen]
currentprod = sentences[idx]
sentences.pop(idx)
similarities = [util.pytorch_cos_sim(choosen, embedding) for embedding in embeddings]
# The max is the most similar product
idx = similarities.index(max(similarities))

currentRating = circ[circ['Product_number'] == choosenID]['Product_circularity']
suggestedRating = circ[circ['Product_number'] == IDS[idx]]['Product_circularity']
# While the suggestion isnt more sustainable than the product itself
# Look for an alternative
while(len(suggestedRating) <= len(currentRating)):
    embeddings = [arr for arr in embeddings if arr != embeddings[idx]]
    sentences.pop(idx)
    similarities.pop(idx)
    idx = similarities.index(max(similarities))
    currentRating = circ[circ['Product_number'] == choosenID]['Product_circularity']
    suggestedRating = circ[circ['Product_number'] == IDS[idx]]['Product_circularity']
print(f"the current product rating is {currentRating} while the suggested product rating is {suggestedRating}")
print(f"the current product is {currentprod} while the suggested product is {sentences(idx)}")
```
