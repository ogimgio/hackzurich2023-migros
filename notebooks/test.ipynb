```python
def try_parse(x):
    try:
        x= int(float(x))
        return str(x)
    except:
        return str(0)


df_fish = pd.read_csv('Migros_case/Fisch.csv',dtype={'Product_number': 'object'},sep=";")
df_fish.Product_number = df_fish.Product_number.apply(try_parse)
```

```python
#merge fish dataframe on product_name with product csv (id)
product = pd.read_csv('Migros_case/products.csv')
product.id = product.id.astype(str)
df_fish = pd.read_csv('Migros_case/Fisch.csv',dtype={'Product_number': 'object'},sep=";")
df_fish.Product_number = df_fish.Product_number.apply(try_parse)
df_fish = df_fish.rename(columns={'Product_number': 'id',"Product_description":"fish_name"})
new_fish = product.merge(df_fish, on='id', how='left')
new_fish
```

```python
# read product csv 
product = pd.read_csv('Migros_case/sust_score.csv')
# order by bigger sust_score
product = product.sort_values(by=['sust_score'], ascending=False)
# get top 10
product.head(10).reset_index()
```

```python
# get index 22306,37812,34613, i want their product_id
healthy_breakfast_ids = product.iloc[[0,7	,9]].product_id.tolist()
healthy_breakfast_scores = product.iloc[[0,7	,9]].sust_score.tolist()

#take healthy_breakfast_ids from product csv
product = pd.read_csv('Migros_case/products.csv')
```

```python
import ast 
selected_product = product[product.id.isin(healthy_breakfast_ids)]
selected_product["image_url"] = selected_product["image"].apply(lambda x: ast.literal_eval(x)["original"])
selected_product["sust_score"] = healthy_breakfast_scores
selected_product[['id','name','sust_score','image_url']]
```
